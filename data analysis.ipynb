```python
#Filtering positive words
def remove_similar_words(main_file, comparison_files):
    with open(main_file, 'r') as main_file_obj:
        main_words = set(main_file_obj.read().split())

    comparison_words = set()

    for comparison_file in comparison_files:
        with open(comparison_file, 'r') as comparison_file_obj:
            comparison_words.update(comparison_file_obj.read().split())

    # Remove similar words
    remaining_words = main_words.difference(comparison_words)

    # Write the remaining words to a new file
    remaining_file = main_file.replace('.txt', '_remaining.txt')
    with open(remaining_file, 'w') as remaining_file_obj:
        remaining_file_obj.write(' '.join(remaining_words))

main_file_path = 'C:\\Users\\NACHI\\Desktop\\python projects\\master_dictionary\\positive-words.txt'
comparison_files_paths =['C:\\Users\\NACHI\Desktop\\python projects\\stopwords\\StopWords_Names.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Geographic.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_GenericLong.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_DatesandNumbers.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Currencies.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Auditor.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Generic.txt']


remove_similar_words(main_file_path, comparison_files_paths)

#Filtering negative words
def remove_similar_words(main_file, comparison_files):
    with open(main_file, 'r') as main_file_obj:
        main_words = set(main_file_obj.read().split())

    comparison_words = set()

    for comparison_file in comparison_files:
        with open(comparison_file, 'r') as comparison_file_obj:
            comparison_words.update(comparison_file_obj.read().split())

    # Remove similar words
    remaining_words = main_words.difference(comparison_words)

    # Write the remaining words to a new file
    remaining_file = main_file.replace('.txt', '_remaining.txt')
    with open(remaining_file, 'w') as remaining_file_obj:
        remaining_file_obj.write(' '.join(remaining_words))

main_file_path = 'C:\\Users\\NACHI\\Desktop\\python projects\\master_dictionary\\negative-words.txt'
comparison_files_paths =['C:\\Users\\NACHI\Desktop\\python projects\\stopwords\\StopWords_Names.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Geographic.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_GenericLong.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_DatesandNumbers.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Currencies.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Auditor.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Generic.txt']


remove_similar_words(main_file_path, comparison_files_paths)
```

```python
positive_score=[]
negative_score=[]
```

```python
#following function extract article title and text
def extract_article(url):
    article = newspaper.Article(url)
    
    # Download and parse the article
    article.download()
    article.parse()
    
    # Extract the title and text of the article
    title = article.title
    text = article.text
    
    return title, text

#this function tokanizes the words
def tokenize_text(text):
    tokens = word_tokenize(text)
    return tokens

#this function removes stop words from th scraped text
def remove_strings_from_list(strings_list, *txt_files):
    for file_path in txt_files:
        with open(file_path, 'r') as file:
            file_content = file.read()
            strings_list = [string for string in strings_list if string not in file_content]

    return strings_list

#this function counts number of positive and negative words
def count_words_from_file(word_list, file_path):
    word_count = 0
    with open(file_path, 'r') as file:
        file_content = file.read()
        for word in word_list:
            if word in file_content:
                word_count += 1
    return word_count

print("positive words")
for link in df['URL']:
    url=link
    title, text = extract_article(url)
    scraped_data = text
    #print(scraped_data)
    token_list = tokenize_text(scraped_data)
    #print("List of tokens:", token_list)
    strings_list = token_list
    filtered_list = remove_strings_from_list(strings_list,'C:\\Users\\NACHI\Desktop\\python projects\\stopwords\\StopWords_Names.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Geographic.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_GenericLong.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_DatesandNumbers.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Currencies.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Auditor.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Generic.txt')
    #print("Filtered list:", filtered_list)
    word_list = filtered_list
    file_path = "C:\\Users\\NACHI\\Desktop\\python projects\\master_dictionary\\positive-words_remaining.txt" 
    total_count = count_words_from_file(word_list, file_path)
    print("Total count :", total_count )
    positive_score.append(total_count)
    
print("negative words")    
for link in df['URL']:
    url1=link
    title, text = extract_article(url1)
    scraped_data = text
    #print(scraped_data)
    token_list1 = tokenize_text(scraped_data)
    #print("List of tokens:", token_list)
    strings_list1 = token_list1
    filtered_list1= remove_strings_from_list(strings_list1,'C:\\Users\\NACHI\Desktop\\python projects\\stopwords\\StopWords_Names.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Geographic.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_GenericLong.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_DatesandNumbers.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Currencies.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Auditor.txt','C:\\Users\\NACHI\\Desktop\\python projects\\stopwords\\StopWords_Generic.txt')
    #print("Filtered list:", filtered_list)
    word_list1 = filtered_list1
    file_path1 = "C:\\Users\\NACHI\\Desktop\\python projects\\master_dictionary\\negative-words_remaining.txt" 
    total_count1 = count_words_from_file(word_list1, file_path1)
    print("Total count :", total_count1 )
    negative_score.append(total_count1)    
```

```python
#appending positive and negative score into df
df['POSITIVE SCORE']=positive_score
df['NEGATIVE SCORE']=negative_score
df
```

```python
Polarity_Score=[]
for i,j in zip(positive_score,negative_score):
    a=(i-j)/((i+j)+0.000001)
    Polarity_Score.append(a)
Polarity_Score
```

```python
def count_sentences(scraped_data):
    # Tokenize the scraped data into sentences
    sentences = nltk.sent_tokenize(scraped_data)
    
    # Count the number of sentences
    num_sentences = len(sentences)
    
    return num_sentences

sentence_count=[]
for link in df['URL']:
    url2=link
    title, text = extract_article(url2)
    scraped_data = text
    sentence_counting = count_sentences(scraped_data)
    sentence_count.append(sentence_counting)
    print("Number of sentences:", sentence_count)
```

```python
Average_sentence=[]
for i,j in zip (words_after_cleaning,sentence_count):
    a=round(i/j)
    
    Average_sentence.append(a)
    
    print(a)
```

```python
#subjectivity score appended to data frame
df['AVERAGE SENTENCE LENGTH']=Average_sentence
df
```

```python
#following function counts complex words
def count_complex_words(scraped_data):
    # Initialize the CMU Pronouncing Dictionary
    d = cmudict.dict()
    
    words = nltk.word_tokenize(scraped_data)
    
    complex_word_count = 0
    
    for word in words:
        try:
            syllable_count = [len(list(y for y in x if y[-1].isdigit())) for x in d[word.lower()]]
        except KeyError:
            continue
        
        # Check if the word has more than two syllables
        if syllable_count and max(syllable_count) > 2:
            complex_word_count += 1
    
    return complex_word_count

complex_words=[]
for link in df['URL']:
    url2=link
    title, text = extract_article(url2)
    scraped_data = text
    complex_word_count = count_complex_words(scraped_data)
    complex_words.append(complex_word_count)
    print("Count of complex words:", complex_word_count)
```

```python
percentage_complex_words=[]
for i,j in zip (complex_words,words_after_cleaning):
    a=i/j
    percentage_complex_words.append(a)
    print(a)
```

```python
df['PERCENTAGE OF COMPLEX WORDS']=percentage_complex_words
df
```
